```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
plt.plot(women_degrees["Year"], women_degrees["Biology"])
plt.show()
```

```python
men_biology_degree = 100 - women_degrees["Biology"]
plt.plot(women_degrees['Year'], women_degrees['Biology'], c='blue', label='Women')
plt.plot(women_degrees['Year'], men_biology_degree, c='green', label='Men')
plt.legend(loc='upper right')
plt.title('Percentage of Biology Degrees Awarded By Gender')
plt.show()
```

```python
# Chartjunk: unnecessary and confusing representations in charts and graphs. 
# We want to maximize the data-ink ratio: fractional amount of the plotting area dedicated to displaying the data
# Non-data-inks are the elements in the chart that don't display data points, such as tick markers, labels, legends
```

```python
# Axes.tick_params() method: customize the appearance of ticks
fig, ax = plt.subplots()
ax.plot(women_degrees['Year'], women_degrees['Biology'], label='Women')
ax.plot(women_degrees['Year'], men_biology_degree, label='Men')

ax.tick_params(top='off', bottom='off', left="off", right="off")
ax.set_title('Percentage of Biology Degrees Awarded By Gender')
ax.legend(loc="upper right")

for key, spine in ax.spines.items():
    spine.set_visible(False)

plt.show()
# By removing the tick marks and spines, we improve the data visibility. 
```

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig2 = plt.figure(figsize=(12, 12))
cb_dark_blue =  (0/255, 107/255, 164/255)
orange = (255/255, 128/255, 14/255)

for sp in range(0,4):
    ax20 = fig2.add_subplot(2,2,sp+1)
    ax20.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax20.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c=orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax20.set_xlim(1968, 2011)
    ax20.set_ylim(0, 100)
    ax20.set_title(major_cats[sp])
    ax20.tick_params(bottom='off',top='off',right='off',left='off')
    ax20.legend(loc='upper right')
    
plt.show()
```

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig3 = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax30 = fig3.add_subplot(1,6,sp+1)
    ax30.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax30.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax30.set_xlim(1968, 2011)
    ax30.set_ylim(0,100)
    ax30.set_title(stem_cats[sp])
    ax30.tick_params(bottom="off", top="off", left="off", right="off")

plt.legend(loc='upper right')
plt.show()
```

```python
fig4 = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax40 = fig4.add_subplot(1,6,sp+1)
    ax40.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax40.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=orange, label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax40.set_xlim(1968, 2011)
    ax40.set_ylim(0,100)
    ax40.set_title(stem_cats[sp])
    ax40.tick_params(bottom="off", top="off", left="off", right="off")
    if sp == 0:
        ax40.text(2005, 87, 'Men')
        ax40.text(2002, 8, 'Women')
    elif sp == 5:
        ax40.text(2005, 62, 'Men')
        ax40.text(2001, 35, 'Women')

plt.legend(loc='upper right')
plt.show()
```
